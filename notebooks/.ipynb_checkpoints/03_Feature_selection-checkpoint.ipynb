```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso , LogisticRegression
from sklearn.linear_model import LassoCV
from sklearn.feature_selection import SelectFromModel
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from statsmodels.stats.outliers_influence import variance_inflation_factor


df = pd.read_csv("../data/interim/data.csv")
```

```python
X = df.drop("LOAN_DEFAULT" , axis = 1)
y = df["LOAN_DEFAULT"]
X_train , X_val , y_train , y_val = train_test_split(X, y, test_size = 0.2, random_state=7)
# X_train_sub_set , X_vald_sub_set , y_train_sub_set , y_vald_sub_set = train_test_split(x[0:1000], y[0:1000] , test_size = 0.2, random_state=42)
```

```python
# scaling data
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_val_scaled = ss.transform(X_val)
```

Our approach is to find the best features that can be cascaded on all models using various techniques like Lasso, Meta transformer of SelectFromModel and Linear Discriminant Analysis; in addition, our business intuition.

## Scale data using z score

```python
# scaling data
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_val_scaled = ss.transform(X_val)
```

## Lasso
<a id='techniques'></a>
<a href="#">Back to top</a>

```python
# define lasso model
lasso = LassoCV(cv=5,n_jobs=-1)
lasso.fit(X_train_scaled, y_train)
# coef of lasso sorted
pd.DataFrame(lasso.coef_, index=X.columns).sort_values(ascending=False, by=[0])
```

## SelectFromModel (Meta-transformer)
<a href="#">Back to top</a>

```python
# select from model where C=0.01 yields more than 6 features

sfm = SelectFromModel(LogisticRegression(C=0.01))
sfm.fit(X_train_scaled, y_train)
selected_feat = X_train.columns[(sfm.get_support())]
print('Total features: {}'.format((X_train.shape[1])))
print('Selected features: {}'.format(len(selected_feat)))
print("*"*50)
print(list(selected_feat))
```

## Linear Discriminant Analysis
<a href="#">Back to top</a>

```python
# define lda
lda = LinearDiscriminantAnalysis()
lda.fit(X_train_scaled, y_train)
pd.DataFrame(lda.coef_[0], index=X.columns).sort_values(ascending=False, by=[0])
```

> Output features sorted based on best coeffients

## Using Variance inflation factor to detect multicollinarity
<a id='vif'></a>
<a href="#">Back to top</a>

In this technique, we are trying to regress each feature against all features. A great rule of thumb is to drop features that outcome of VIF is greater than 10. 

Formula:
$$ VIF = 
\frac{1}{1-R{}^2}$$

```python
# chosen columns out of the above experiments
colname = ['DISBURSED_AMOUNT', 'ASSET_COST',
       'PRI_ACTIVE_ACCTS', 'PRI_OVERDUE_ACCTS',
       'PRI_DISBURSED_AMOUNT', 'PRIMARY_INSTAL_AMT',
       'NEW_ACCTS_IN_LAST_SIX_MONTHS',
       'DELINQUENT_ACCTS_IN_LAST_SIX_MONTHS', 'NO_OF_INQUIRIES',
       'LOANEE_DOB_DAYS', 'DISBURSAL_DATE_DAYS',
       'LTA', 'EMPLOYMENT_TYPE_Self employed',
       'PERFORM_CNS_DESC_Low', 'PERFORM_CNS_DESC_Medium',
       'PERFORM_CNS_DESC_High', 'PERFORM_CNS_DESC_Very high']
```

```python
# Calculate VIF for every feature
pd.DataFrame([variance_inflation_factor(X[colname], i)
                          for i in range(len(X[colname].columns))], index=X[colname].columns)
```
